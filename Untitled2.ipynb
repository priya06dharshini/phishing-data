```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("dataset.xlsx")

plt.subplot(2,2,1)
plt.plot(df['num_words'], df['num_unique_words'], color='red')
plt.xlabel("num_words")
plt.ylabel("num_unique_words")

plt.subplot(2,2,2)
plt.pie(df['num_unique_words'][:5], labels=df['num_words'][:5])

plt.subplot(2,2,3)
plt.bar(df['num_words'][:10], df['num_unique_words'][:10], color='red')
plt.xlabel("num_words")
plt.ylabel("num_unique_words")

plt.subplot(2,2,4)
plt.scatter(df['num_words'], df['num_unique_words'], color='red')
plt.xlabel("num_words")
plt.ylabel("num_unique_words")

plt.tight_layout()
plt.savefig("dashboard.png",dpi=300,bbox_inches='tight')
plt.show()
```
